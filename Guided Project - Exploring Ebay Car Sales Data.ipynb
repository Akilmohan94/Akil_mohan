```python
autos.brand.value_counts(normalize=True)
```

```python
brand_prices = {}
top_10 = autos.brand.value_counts(normalize = True).index[:10]
```

```python
top_10
```

```python
for brand in top_10: 
    group = autos[autos.brand == brand ]
    brand_prices[brand] = int(group.price.mean())
    
```

```python
top_10
```

```python
brand_prices
```

```python
sorted(brand_prices, key=brand_prices.get)
 
```

- BMW, AUDI and Mercedes are more expensive. 
- Volkswagen belongs to the mid range. 
- Ford and Opel are less expensive

```python
brand_mileage = {} 
for brand in top_10: 
    grp = autos[autos.brand == brand]
    brand_mileage[brand] = int(grp.odometer_km.mean())
```

```python
brand_mileage
```

```python
mil = pd.Series(brand_mileage)
```

```python
price = pd.Series(brand_prices
         )
```

```python
final = pd.concat([mil,price],axis =1)
```

```python
final
```

```python
final.rename({0:'Mean_Mileage',1:'Mean_price'},axis =1,inplace = True)
```

```python
final
```

```python
final.sort_values(by = 'Mean_price')
```

Among the top 10 brands, we can see that theres not much difference with respect to Mean of the mileage, but in terms of their pricings they can be separated into 3 classes that we have mentioned above.
