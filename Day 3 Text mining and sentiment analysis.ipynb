### Format the data into right dataframe

```python
tweets_user = api.user_timeline('msdhoni',count=100)
len(tweets_user)
```

```python
tweets = pd.DataFrame(columns=['id','created_at','likes','retweets','text','source'])
for tweet in tweets_user:
    curr_tweet = {'id':tweet.id,
                 'created_at':tweet.created_at,
                 'likes':tweet.favorite_count,
                 'retweets':tweet.retweet_count,
                 'text':tweet.text,
                 'source':tweet.source}
    tweets = tweets.append(curr_tweet , ignore_index=True)
tweets['created_at'] = tweets['created_at'].astype(str)
tweets['created_at'] = pd.to_datetime(tweets['created_at'],format = '%Y-%m-%d %H:%M:%S')
tweets.to_csv('csk.csv',index=False)
```
